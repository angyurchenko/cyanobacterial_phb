```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_names = ['t', 'PHB']
Scytonema_geitleri_Bharadwaja = pd.DataFrame({15: 2.35,
                                30: 3.18}.items(), columns=col_names)
Nostoc_muscorum_NCCU442 = pd.DataFrame({15: 1.8,
                          20: 3.2,
                          25: 5.5,
                          30: 6.5,
                          40: 2.2}.items(), columns=col_names)
Synechocystis_PCC6803 = pd.DataFrame({16: 2.35,
                        20: 3.31,
                        24: 3.33,
                        28: 4.55,
                        32: 4.23,
                        36: 3.86}.items(), columns=col_names)
```

```python
plt.figure(figsize=(9, 7), dpi=300)

sns.lineplot(data= Scytonema_geitleri_Bharadwaja, x='t', y='PHB', 
             marker='o', markersize=5, 
             linestyle='--', label='Scytonema geitleri Bharadwaja')

sns.lineplot(data=Nostoc_muscorum_NCCU442, x='t', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Nostoc muscorum NCCU-442')

sns.lineplot(data=Synechocystis_PCC6803, x='t', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Synechocystis PCC 6803')

plt.legend()
plt.ylabel('PHB, %')
plt.xlabel('Temperature, $^\circ$C')
#plt.savefig('temperature.png')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8,8))

sns.barplot(data= Scytonema_geitleri_Bharadwaja, x='t', y='PHB', 
            color='lightgreen', label='Scytonema geitleri Bharadwaja', ax=ax1)

sns.barplot(data=Nostoc_muscorum_NCCU442, x='t', y='PHB',
            color='seagreen', label='Nostoc muscorum NCCU-442',ax=ax2)

sns.barplot(data=Synechocystis_PCC6803, x='t', y='PHB',
            color='darkgreen', label='Synechocystis PCC 6803',ax=ax3)

#plt.legend()
plt.xlabel('Temperature, $^\circ$C')
ax1.set_title('Scytonema geitleri Bharadwaja', fontsize=15)
ax2.set_title('Nostoc muscorum NCCU-442', fontsize=15)
ax3.set_title('Synechocystis PCC 6803', fontsize=15)

ax1.set_xlabel('')
ax2.set_xlabel('')
ax3.set_xlabel('Temperature, $^\circ$C', fontsize=15)

ax1.set_ylabel('PHB, %', fontsize=15)
ax2.set_ylabel('PHB, %', fontsize=15)
ax3.set_ylabel('PHB, %', fontsize=15)

plt.subplots_adjust(hspace=.8)
plt.savefig('temperature.png')
```
