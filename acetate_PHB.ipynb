```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_names = ['acetate', 'PHB']
Scytonema_geitleri_Bharadwaja = pd.DataFrame({0 : 2.82, 
                                 10: 6.01, 
                                 20: 6.71, 
                                 30: 7.12, 
                                 40: 5.90, 
                                 50: 5.73}.items(), columns=['acetate', 'PHB'])

Aulosira_fertilissima = pd.DataFrame({0: 6.4,
                        33: 8.5,
                        50: 14.5,
                        83: 14.5,
                        167: 12.7}.items(), columns=col_names)

Synechocystis_PCC6803_N_def = pd.DataFrame({0: 9.5,
                              33: 11.9,
                              67: 14.6,
                             100: 11.5}.items(), columns=col_names)
Synechocystis_PCC6803_P_def = pd.DataFrame({0: 11.2,
                              33: 15.9,
                              67: 25.7,
                              100: 21.5}.items(), columns=col_names)

Spirulina_platensis_UMACC159 = pd.DataFrame({0: 3,
                               83: 7}.items(), columns=col_names)
Spirulina_platensis_UMACC161 = pd.DataFrame({0: 1,
                               83: 10}.items(), columns=col_names)
Spirulina_platensis_UMACC162 = pd.DataFrame({0: 1,
                               83: 8}.items(), columns=col_names)
```

```python
plt.figure(figsize=(9, 7), dpi=300)

sns.lineplot(Scytonema_geitleri_Bharadwaja.acetate, Scytonema_geitleri_Bharadwaja.PHB, 
             marker='o', markersize=5, 
             linestyle='--', label='Scytonema geitleri Bharadwaja')

sns.lineplot(Aulosira_fertilissima.acetate, Aulosira_fertilissima.PHB, 
             marker='o', markersize=5, 
             linestyle='--', label='Aulosira fertilissima')

sns.lineplot(Synechocystis_PCC6803_N_def.acetate, Synechocystis_PCC6803_N_def.PHB, 
             marker='o', markersize=5,
             linestyle='--', label='Synechocystis PCC6803+N-deficiency')

sns.lineplot(Synechocystis_PCC6803_P_def.acetate, Synechocystis_PCC6803_P_def.PHB, 
             marker='o', markersize=5,
             linestyle='--', label='Synechocystis PCC6803+P-deficiency')

sns.lineplot(Spirulina_platensis_UMACC159.acetate, Spirulina_platensis_UMACC159.PHB,
             marker='o', markersize=5,
             linestyle='--', label='Spirulina platensis UMACC 159')

sns.lineplot(Spirulina_platensis_UMACC161.acetate, Spirulina_platensis_UMACC161.PHB,
             marker='o', markersize=5,
             linestyle='--', label='Spirulina platensis UMACC 161')

sns.lineplot(Spirulina_platensis_UMACC162.acetate, Spirulina_platensis_UMACC162.PHB,
             marker='o', markersize=5,
             linestyle='--', label='Spirulina platensis UMACC 162')

plt.legend()
plt.ylabel('PHB, %')
plt.xlabel('acetate, mM')
#plt.savefig('acetate.png') <-- in case you need to save your masterpiece 
```
