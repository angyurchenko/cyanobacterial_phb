```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""Creating all necessary df to plot them"""

col_names=['pH', 'PHB']

Scytonema_geitleri_Bharadwaja = pd.DataFrame({5.5: 0.5,
                                7.5: 2.6,
                                8.5: 3.98,
                                10.5: 1.8}.items(), columns=col_names)

Nostoc_muscorum_NCCU442 = pd.DataFrame({5: 1.2,
                          5.5: 1.5,
                          6: 2.3,
                          6.5: 4.2,
                          7: 4.9,
                          7.5: 7.6,
                          8: 6.8,
                          8.5: 6.5,
                          9: 6.2,
                          9.5: 5.8,
                          10: 3.1}.items(), columns=col_names)

Synechocystis_PCC6803 = pd.DataFrame({5.5: 1.31,
                        6.5: 1.33,
                        7.5: 4.22,
                        8.5: 4.55,
                        9.5: 2.86,
                        10.5: 2.31}.items(), columns=col_names)

Spirulina_platensis_UMACC159 = pd.DataFrame({4: 0,
                               6: 1.9,
                               7: 2.0,
                               9: 7.1,
                               11: 4.4,
                               12: 4.3}.items(), columns=col_names)

Spirulina_platensis_UMACC161 = pd.DataFrame({4: 2.0,
                               6: 4.1,
                               7: 5.7,
                               9: 10.1,
                               11: 5.2,
                               12: 5.1}.items(), columns=col_names)

Spirulina_platensis_UMACC162 = pd.DataFrame({4: 0,
                               6: 3.1,
                               7: 5.2,
                               9: 7.8,
                               11: 6.3,
                               12: 5.8}.items(), columns=col_names)
```

```python
"""Plotting dfs"""

plt.figure(figsize=(9, 7), dpi=300)

sns.lineplot(data= Scytonema_geitleri_Bharadwaja, x='pH', y='PHB', 
             marker='o', markersize=5, 
             linestyle='--', label='Scytonema geitleri Bharadwaja')

sns.lineplot(data=Nostoc_muscorum_NCCU442, x='pH', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Nostoc muscorum NCCU-442')

sns.lineplot(data=Synechocystis_PCC6803, x='pH', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Synechocystis PCC 6803')

sns.lineplot(data=Spirulina_platensis_UMACC159, x='pH', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Spirulina platensis UMACC-159')

sns.lineplot(data=Spirulina_platensis_UMACC161, x='pH', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Spirulina platensis UMACC-161')

sns.lineplot(data=Spirulina_platensis_UMACC162, x='pH', y='PHB',
            marker='o', markersize=5, 
            linestyle='--', label='Spirulina platensis UMACC-162')

plt.legend()
plt.ylabel('PHB, %')
plt.xlabel('pH')
#plt.savefig('pH.png')
```
